```python
# Histogram of koi_period
plt.figure(figsize=(10, 6))
plt.hist(df['koi_period'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of koi_period')
plt.xlabel('koi_period')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Scatter plot of koi_depth vs koi_duration
plt.figure(figsize=(10, 6))
plt.scatter(df['koi_depth'], df['koi_duration'], color='green', alpha=0.5)
plt.title('Scatter plot of koi_depth vs koi_duration')
plt.xlabel('koi_depth')
plt.ylabel('koi_duration')
plt.grid(True)
plt.show()
```

```python
# Box plot of koi_impact
plt.figure(figsize=(8, 6))
plt.boxplot(df['koi_impact'].dropna(), vert=False)
plt.title('Box plot of koi_impact')
plt.xlabel('koi_impact')
plt.grid(True)
plt.show()

```

```python
# Bar plot of koi_disposition
plt.figure(figsize=(8, 6))
df['koi_disposition'].value_counts().plot(kind='bar', color='orange')
plt.title('Bar plot of koi_disposition')
plt.xlabel('Disposition')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
#Heatmap
numeric_data=df.select_dtypes(include=['int64','float64'])
correlation_matrix = numeric_data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pie Chart of koi_disposition
plt.figure(figsize=(8, 8))
df['koi_disposition'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Pie Chart of koi_disposition')
plt.ylabel('')
plt.show()
```

```python
# Pair Plot
sns.pairplot(df[['koi_period', 'koi_duration', 'koi_depth', 'koi_impact', 'koi_prad']], kind='reg', diag_kind='kde')
plt.suptitle('Pair Plot', y=1.02)
plt.show()

```
